```python
import pandas as pd
# contains data for all countries and their GDP in 2014
gdp = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2014_world_gdp_with_codes.csv')
gdp.head()
```

```python
# Data contains all loans sorted by country
countrycount = pd.read_csv('/Users/jiaminlim93/Desktop/GA Data Science Course/project/Data Science Project_Jiamin_/countrycount.csv')
#countrycount.head(20)
countrycount.head()
```

```python
# Cleaning data so country names match with package
countrycount.set_value(11,'country','Bolivia, Plurinational State of')
countrycount.set_value(16,'country','Viet Nam')
countrycount.set_value(20,'country','Tanzania, United Republic of')
countrycount.set_value(31,'country','Palestine, State of')
countrycount.set_value(47,'country','Congo, The Democratic Republic of the')
countrycount.set_value(59,'country',"Lao People's Democratic Republic")
countrycount.set_value(60,'country','Congo')
countrycount.set_value(67,'country',"Moldova, Republic of")
countrycount.set_value(68,'country',"Myanmar")
countrycount.set_value(84,'country',"Saint Vincent and the Grenadines")

countrycount = countrycount.drop(countrycount.index[[57,72,86]]) #removing Kosovo & Cote d'Ivoire & Gaza
countrycount.tail(50)
```

```python
import pycountry
list(pycountry.countries)
```

```python
# Adding on country codes to the country count file
code = []

# For each row in the column
for row in countrycount['country']:
    country = pycountry.countries.get(name=row)
    code.append(country.alpha_3)

countrycount['CODE'] = code
countrycount.to_csv('countrycountwithcode.csv')
#countrycount.head()
```

```python
# Creating a full dataset so all the map lines come up in the choropleth
total = pd.merge(gdp, countrycount, on='CODE', how='left')
total = total.drop(total.columns[[3,4]], axis=1)
total = total[['COUNTRY', 'CODE', 'GDP (BILLIONS)', 'count']]
total['count'].fillna(0,inplace=True)
total.head(10)
total.to_csv('total.csv')
```
