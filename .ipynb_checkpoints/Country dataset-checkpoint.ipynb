```python
import pandas as pd
# Importing list of country names
list = pd.read_csv('/Users/jiaminlim93/Desktop/GA Data Science Course/project/Data Science Project_Jiamin_/countrycount.csv')
#countrycount.head(20)
list = list.drop(['Unnamed: 0','count'],axis=1)
```

```python
len(list)
```

```python
# Part 1: Importing ISO-3 codes
# Cleaning data so country names match with package
list['country2'] = list['country']
list.set_value(11,'country2','Bolivia, Plurinational State of')
list.set_value(16,'country2','Viet Nam')
list.set_value(20,'country2','Tanzania, United Republic of')
list.set_value(31,'country2','Palestine, State of')
list.set_value(47,'country2','Congo, The Democratic Republic of the')
list.set_value(59,'country2',"Lao People's Democratic Republic")
list.set_value(60,'country2','Congo')
list.set_value(67,'country2',"Moldova, Republic of")
list.set_value(68,'country2',"Myanmar")
list.set_value(84,'country2',"Saint Vincent and the Grenadines")

list = list.drop(list.index[[57,72,86]]) #removing Kosovo & Cote d'Ivoire & Gaza
list = list.reset_index(drop=True)
list
```

```python
import pycountry
code = []

# For each row in the column
for row in list['country2']:
    country = pycountry.countries.get(name=row)
    code.append(country.alpha_3)

list['a-3'] = code
len(list)
```

```python
import numpy as np
# Manually add Kosovo & Cote d'Ivoire & Gaza
#list.loc = np.array(['Kosovo', 'XKX'])  # adding a row
list.loc[90] = ['Kosovo','Kosovo', 'XKX']
list.loc[91] = ["Cote d'Ivoire","Cote d'Ivoire",'CIV']
list.loc[92] = ["Gaza","Gaza", 'N/A']
```

```python
list
```

```python
# Part 2: Importing continent data 
# Note: for countries that are in Asia and Europe, they are labelled as being in europe (this was altered in the source Wiki file)
Wikidata = pd.read_csv('/Users/jiaminlim93/Desktop/GA Data Science Course/project/Data Science Project_Jiamin_/Wiki_country.csv')
Wikidata.head()

total = pd.merge(list, Wikidata, on='a-3', how='left')
total = total[['country', 'C','a-2','a-3']]
#total.loc[total['C'] == 'AS']
```

```python
#Filling in missing data for Gaza and Kosovo
total.set_value(90,'C','EU')
total.set_value(92,'C','AS')
#total.C.value_counts()
total
```

```python
total.to_csv('countrydataset.csv')
```
